And now we want to check want happened to 42

```r
options(repr.plot.width=16, repr.plot.height=8)
layout(matrix(1:2,nrow=1))
plot(HEIGHT,type = 'n')
text(IDNO,y=HEIGHT,labels = IDNO)
plot(ADIPOSITY[-42],WEIGHT[-42]/HEIGHT[-42]^2,type = 'n')
text(ADIPOSITY[-42],WEIGHT[-42]/HEIGHT[-42]^2,IDNO)
layout(1)
```

We can see the height of 42 is quite far from the majority and is unreasonably low, we think it is a wrong record.

Recalculate weight based on ADIPOSITY and WEIGHT

```r
bodyfat[42,'HEIGHT'] <- sqrt(bodyfat[42,'WEIGHT']/bodyfat[42,'ADIPOSITY']*703)
bodyfat[162,'HEIGHT'] <- sqrt(bodyfat[162,'WEIGHT']/bodyfat[162,'ADIPOSITY']*703) #Recalculate weight based on ADIPOSITY and WEIGHT 
bodyfat[220,'HEIGHT'] <- sqrt(bodyfat[220,'WEIGHT']/bodyfat[220,'ADIPOSITY']*703) #Recalculate weight based on ADIPOSITY and WEIGHT 
```

We will check the outliers 39,41,212 to see if there is any mistake

```r
set.seed(2)
x <- sample(1:252,22)
y <- x[!x %in% c(48,76,96,182)]

subdata <- bodyfat[unique(c(39,41,212,y)),c('IDNO',"ADIPOSITY","BODYFAT","CHEST","NECK",'HEIGHT','WRIST')]
stars(subdata[,c("ADIPOSITY","BODYFAT","CHEST","NECK",'HEIGHT','WRIST')],
      nrow=5,ncol=5,labels=subdata$IDNO)
```

The first 3 are the suspects, each hexagon represents the person's ADIPOSITY,BODYFAT,CHEST,NECK,HEIGHT,WRIST. We can see 39 is a big guy is every way so the data collected are consistent. 212 has the same 'shape' with many other persons like 203 so we think it is also a correct record. 41 may be a problem since it doesn't look like others. But the left lower diagonal represents 'HEIGHT' which is already crosschecked before.
